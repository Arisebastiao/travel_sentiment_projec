On this second part I will analyze the probabibilyt of being an expat and be more likely to answer to a certain way.

P(A|B) = P(A  B) / P(B)

P(A|B) = probability of A occurring, given B occurs
P(A  B) = probability of both A and B occurring
P(B) = probability of B occurring

 P(responding that will be ok to travel after 70% get vaccinanted | While being expat)

```python
# first let me get the universe of expats

travel_sentiment["all_axpats"] = travel_sentiment['Expat?'][travel_sentiment['Expat?'] == "Yes"]
```

```python
#Than lets get the universe of people that said
travel_sentiment["majority_vacc"] = travel_sentiment['when do you think will be ok to travel again ?'][travel_sentiment['when do you think will be ok to travel again ?'] == 'After atleast 70% of the population of your destination had the vaccine.']
```

```python
travel_sentiment["always"] = travel_sentiment['when do you think will be ok to travel again ?'][travel_sentiment['when do you think will be ok to travel again ?']== 'Its always OK to travel.']
```

P(A) = responding that will be ok to travel after 70% get vaccinanted 
P(B) =  While being expat
P(A|B = responding that will be ok to travel when 70% get vaccinated while being expat

We put will put the formula working:
P(A|B) = P(A  B) / P(B)

```python
# 38(events) / 4 (possible outcomes) probabilidade of responding that will be ok to travel after 70% of the pop gets the vaccine

prob_a = 4 / 38
prob_a
```

```python
# 38 / 16 probabilidade of being expat

prob_b = 16 / 38
prob_b
```

```python
# calculating the intersection between prob a and b
intersection = 6 / 38
intersection
```

```python
# P(A|B) = P(A  B) / P(B)

print("The probabilidade of responding that will be ok to travel once 70% of the peoplee gets vaccinated is:", intersection / prob_b)
```
